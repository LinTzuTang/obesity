```python
def integer_encode_table():
    snp_uni = ['00', 'AA', 'AC', 'AG', 'AT', 'CA', 'CC', 'CG', 'CT', 'GA', 'GC','GG', 'GT', 'TA', 'TC', 'TG', 'TT']
    integer = list(range(len(snp_uni)))
    snp_uni
    table = {}
    for index,key in enumerate(snp_uni):
        table[key] = integer[index]
    return table

# generate SNP encoding table(dict)
def SNP_encode_table():
    table=integer_encode_table()
    snp_table={}
    x=[]
    for key,value in table.items():
        if value == 0:
            x.append(value)
            snp_table[key] = value
            continue
        else:
            for other_key in list(snp_table.keys()):
                if set(key) == set(other_key): 
                    snp_table[key]=snp_table[other_key]
                    value = snp_table[other_key]
                continue
            x.append(value)
            while value-1 not in x:
                x[-1] = value-1
                value = value -1       
            snp_table[key]=value
    return snp_table
```

```python
# customized activation function
from tensorflow.python.keras.engine.base_layer import Layer
from tensorflow.python.keras.utils import tf_utils


class CustomizedAct(Layer):
    def __init__(self,min_value,max_value,alpha):
        super().__init__()
        self._min_value = min_value
        self._max_value = max_value
        self._alpha = alpha

    def get_config(self):
        base_config = super().get_config()
        base_config.update({'alpha':self._alpha,'min':self._min_value,'max':self._max_value})
        return base_config
        

    def call(self, x, training=True):
        alpha = self._alpha if training else 0
        x_1 = tf.nn.leaky_relu(x-self._min_value,alpha)+self._min_value
        x_2 = -tf.nn.leaky_relu(-x+self._max_value,alpha)+self._max_value
        return x_1 + x_2 - x
    
    
    @tf_utils.shape_type_conversion
    def compute_output_shape(self, input_shape):
        return input_shape
        
```

```python
# model architecture
import tensorflow.keras
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Input,Flatten,BatchNormalization
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras import Model
from tensorflow.keras import optimizers
# n_cols = train_data.shape[1]
def dense_model(n_cols):
    #get number of columns in training data
    #n_cols = train_data.shape[1]
    # Input - Layer
    input_ = Input(shape=(n_cols))
    #x = Dropout(0.99)(input_)
    x = Embedding(11, 8)(input_)
    x = Flatten()(x)
    x = Dense(32, activation= LeakyReLU())(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(32, activation= LeakyReLU())(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(32, activation= LeakyReLU())(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    # Output- Layer
    x = Dense(1, activation=None)(x)
    output = CustomizedAct(10,60,0.1)(x)
    
    model = Model(input_, output)
    model.compile(optimizer = optimizers.Adam(lr=1e-1),
              loss='MSE')
    return model
```
